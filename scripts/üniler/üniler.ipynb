```python
import pandas as pd

# CSV veya txt dosyanız varsa
df = pd.read_csv(r"C:\Users\E-SPOR\Desktop\egitim2\egitim2\scripts\üniler\eleman_bolumler_modified.csv")
```

```python
df.columns = [
    "Bölüm", "Üniversite", "Yıl", "Sıralama", "TabanPuan", 
    "Kontenjan", "UniKodu", "PuanTürü"
]
```

```python
df
```

```python
# Yalnızca önemli sütunları pivotlamak
df_wide = df.pivot_table(
    index=["Bölüm", "Üniversite", "UniKodu", "PuanTürü"], 
    columns="Yıl",
    values=["Sıralama", "TabanPuan", "Kontenjan"],
    aggfunc='first'  # Aynı hücrede birden fazla değer varsa ilkini al
)

# Kolon isimlerini düzleştirelim
df_wide.columns = [f"{val}_{col}" for val, col in df_wide.columns]
df_wide = df_wide.reset_index()
```

```python
df_wide
```

```python
df_wide_sorted = df_wide.sort_values(by="TabanPuan_2024", ascending=False)
df_wide_sorted
```

```python
df_wide_sorted['Bölüm'] = df_wide_sorted['Bölüm'].str.replace('Taban Puanları 2025','')
df_wide_sorted
```

```python
df_wide_sorted.to_json(r"C:\Users\E-SPOR\Desktop\egitim2\egitim2\scripts\üniler\üniler.json", orient='records', force_ascii=False)
```
